```python
import pandas as pd

df = pd.read_csv('combined_wine_quality.csv')
df
```

```python
# New Features

# Alcohol Density
df['alcohol_density'] = df['alcohol'] * df['density']
# Sulfur Dioxide Difference
df['non-free sulfur dioxide'] = df['total sulfur dioxide'] - df['free sulfur dioxide']
# Total Acidity
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
# Acidity Ratio
df['acidity_ratio'] = df['fixed acidity'] / df['total_acidity']
# Alcohol to pH Ratio
df['alcohol_pH_ratio'] = df['alcohol'] / df['pH']
# Alcohol to Sulfur Dioxide Ratio
df['alcohol_sulfur_dioxide_ratio'] = df['alcohol'] / df['total sulfur dioxide']

df
```

```python
# write to CSV
df.to_csv('combined_and_preprocessed_data.csv', index=False)
```
